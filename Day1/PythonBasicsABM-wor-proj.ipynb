```python
# import modules
import numpy
from random import randint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

env_size = 100 # environment size

env = numpy.zeros((env_size,env_size))  # creates an environment of size 1:100 x 1:100 - all values in the arrray are currently zero
# sets xy range where there is food
env[46:50,46:50]=10
env[20:22,20:22]=10
env[60:62,60:62]=10

# start positions in bottom left
xpos = [0]*100
ypos = [0]*100

# random start locations
"""xpos = [randint(0,99) for r in range(100)]
ypos = [randint(0,99) for r in range(100)]"""

agent_eat = [0]*100

# copy of array for plotting time at 0
env0 = numpy.array(env, copy=True)

#plt.imshow(env)
```

```python
for time in range(100): # no of time steps
    for agent in range(100): # no of agents
        food = env[xpos[agent],ypos[agent]] # get value of food at agentn location
        if food == 0: # no food, move
            xpos[agent]= xpos[agent]+randint(-1,5) # possible x movements
            ypos[agent]= ypos[agent]+randint(-1,5) # possible y movements
        else: # food, count
            agent_eat[agent] += 1
            env[xpos[agent],ypos[agent]] = env[xpos[agent],ypos[agent]]-1
            
        ##### error checking - make sure it does not exceed the limits of the environment
        if xpos[agent] >= (env_size-1):
            xpos[agent] = 1
        if xpos[agent] < 1:
            xpos[agent] = (env_size-1)
        if ypos[agent] >= (env_size-1):
            ypos[agent] = 1
        if ypos[agent] < 1:
            ypos[agent] = (env_size-1)

        #plot at time
        if time == 25:
            env25 = numpy.array(env, copy=True)
            #plt.imshow(env)
            plt.scatter(xpos,ypos, color='red', label='25 seconds', marker='x')
        if time == 50:
            env50 = numpy.array(env, copy=True)
            #plt.imshow(env)            
            plt.scatter(xpos,ypos, color='blue', label='50 seconds', marker='x')
        if time == 75:
            env75 = numpy.array(env, copy=True)
            #plt.imshow(env)            
            plt.scatter(xpos,ypos, color='green', label='75 seconds', marker='x')
```
