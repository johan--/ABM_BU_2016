# Making multiple agents.

To have more agents, it would be possible to have more variables.

agent1_xpos = 10

and so on. But this becomes difficult to manage. Programing languages not only have variables (which I am considering as storing a single number), but also have 'arrays' - these can store lots of numbers (or words, or letters, or even locations of other files) in order. While there are other ways to do this - I like to create agent based models where I have an array of x coordiantes and an array of y cooridinates. I can then refer to the position of any agent by using an index term to indicate where they are in the array.

```python
xpos = [10,12,15,16]
ypos = [10,11,13,12]

# to find the position of agent 3 I can then run this command

print(xpos[2], ypos[2])# it is [2] because the first agent will have the 'index' of zero
```

```python
xpos = [10,12,15,16]
ypos = [10,11,13,12]

model_time = 3 # this is a variable which you may want to change - having it at the begining of the code like this makes it easy to do


for timestep in range(model_time): # the model will run for however long is specificed by the model_time variable - initially four timesteps
    for agent in range(3): # there are four agents - zero to 3
        # note that this line is now indented twice - anything here runs in the agent loop - but the agent loop itself runs every timestep
        xpos[agent] = xpos[agent] + 1 # as it is an array, you need to add the index term - in this case we use that value of 'agent' from the loop to indicate which agent we want to look at
        ypos[agent] = ypos[agent] + 1
        

print(xpos, ypos)
```

```python
from random import randint # importing the generating random interger function from the random number package

xpos = 10
ypos = 10  # setting intial coordiantes

heading = 4 # heading is simply a number between 1 and 8, which I will use to indicate a direction for the agent

for timestep in range(10):
    change_direction = randint(-1,1) # creates a variable with value -1, 0, or 1
    heading = heading + change_direction # altering the heading based on the random number
    
    ### this next section of code is 'error checking' I need 'heading' to stay between 1 and 8 (one of the 8 possible directions including diagonal)
    # this value between 1 and 8 is vital in the next bit of code - so I make sure it will always be in this range here
    if heading > 8:
        heading = heading - 8
    if heading < 1:
        heading = heading + 8
    ## end of the error checking - I now have a direction - up, down, left, right and the diagonals. 
    
    if heading == 1: # note double equals sign - this is the test for equality. Single equals means assignment (x = x+1 is assignment, x == x+1 is incorrect)
        ypos = ypos + 1 # move upwards
    if heading == 2:
        xpos = xpos+1
        ypos = ypos+1 # up and right
    if heading == 3:
        xpos = xpos+1 # right
    if heading == 4:
        xpos = xpos + 1
        ypos = ypos -1
    if heading == 5:
        ypos = ypos - 1
    if heading == 6:
        ypos = ypos - 1
        xpos = xpos -1
    if heading == 7:
        xpos = xpos - 1
    if heading == 8:
        xpos = xpos-1
        ypos = ypos+1
    
    print(xpos,ypos)
    ## note how all of this is indented - so is part of the timetep loop - if you unindent the final line (the print line), 
    #then you will just get the final position, rather than the position each timestep
    
    # bear in mind movement is stochasitic - so if you run this several times you won't get the same results
    
```

```python
%matplotlib inline 

import matplotlib.pyplot as plt

import numpy # functions for dealing with numbers, and especially 2D arrays

env_size = 100 # sets dimesnsions for the size of the 2D array which will represent our environment

env = numpy.random.random((env_size,env_size))  # creates an environment of size 1:100 x 1:100 - 

xpos = 50
ypos = 50

current_food = env[xpos, ypos]

x_move = 0
y_move = 0 # setting up some variables to work out how to move to the best food
best_food = current_food

for x in range(-1,1): 
    for y in range(-1,1):  # nested loops to search for the best food...
        food = env[xpos+x,ypos+y]
        if food > best_food:
            best_food = food
            x_move = x
            y_move = y

print('Best food is', best_food)
print('at location x = ',xpos+x_move)
print('y = ',ypos+y_move)

plt.imshow(env[49:51,49:51])
```
